```python
import pandas as pd
```

```python
df_2005=pd.read_csv("data/checking_data/health/health_2005.CSV",encoding='cp949')
df_2010=pd.read_csv("data/checking_data/health/health_2010.CSV",encoding='cp949')
df_2015=pd.read_csv("data/checking_data/health/health_2015.CSV",encoding='cp949')
df_2020=pd.read_csv("data/checking_data/health/health_2020.CSV",encoding='cp949')
```

```python
df_2005 = df_2005.iloc[:1000,:]
df_2010 = df_2010.iloc[:1000,:]
df_2015 = df_2015.iloc[:1000,:]
df_2020 = df_2020.iloc[:1000,:]
```

```python
df_li=[
    df_2005,
    df_2010,
    df_2015, 
    df_2020
    ]

# df columns 공백 제거
for df in df_li:
    df.columns = [i.replace(" ","") for i in df.columns]
```

```python
using_columns=['기준년도', '시도코드', '성별코드', '연령대코드(5세단위)', '신장(5Cm단위)', '체중(5Kg단위)', '시력(좌)', '시력(우)', 
               '청력(좌)', '청력(우)', '수축기혈압', '이완기혈압', '식전혈당(공복혈당)', '총콜레스테롤', 
               '혈색소', '요단백', '(혈청지오티)AST', '(혈청지오티)ALT', '감마지티피', '흡연상태', 
               '음주여부', '허리둘레', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤', '혈청크레아티닌']
```

```python
df_2005.loc[:,using_columns[:-5]]
df_2010.loc[:,using_columns[:-5]]
df_2015.loc[:,using_columns[:-5]]
df_2020.loc[:,using_columns[:-5]]
```

```python
df_tot=pd.concat([df_2005.loc[:,using_columns[:-5]],
                  df_2010.loc[:,using_columns[:-5]],
                  df_2015.loc[:,using_columns[:-5]]],
                 axis=0)

df_tot['음주여부']=df_tot['음주여부'].map({'N':0,'Y':1,1:1,0:0})
```

```python
check_list = ['신장(5Cm단위)',
       '체중(5Kg단위)', '허리둘레', '시력(좌)', '시력(우)', '청력(좌)', '청력(우)', '수축기혈압',
       '이완기혈압', '식전혈당(공복혈당)', '총콜레스테롤', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤',
       '혈색소', '요단백', '혈청크레아티닌', '(혈청지오티)AST', '(혈청지오티)ALT', '감마지티피', '흡연상태',
       '음주여부']
```

```python
for check in check_list:
    print(f"--{check}--")
    print(df_tot.groupby('기준년도')[check].mean())
```
